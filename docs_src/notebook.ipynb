```

getal = 7
vierkant = getal ** 2
print(f"Het kwadraat van {getal} is {vierkant}.")
```

```

import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [i**2 for i in x]

plt.figure(figsize=(6, 4))
plt.plot(x, y, marker='o')
plt.title("Kwadraat van getallen")
plt.xlabel("Getal")
plt.ylabel("Kwadraat")
plt.grid(True)
plt.show()
```
